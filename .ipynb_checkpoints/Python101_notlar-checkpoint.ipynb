## Karşılaştırma Operatörleri

* Eşittir (==)                 Kullanım: 10==10      sonuç = True
* Eşit değildir (!=)           Kullanım: 10!=9       sonuç = True
* Büyüktür (>)                 Kullanım: 10>5        sonuç = True
* Küçüktür (<)                 Kullanım: 10<5        sonuç = False
* Büyük ya da eşittir (>=)     Kullanım: 10>=9       sonuç = True
* Küçük ya da eşittir (<=)     Kullanım: 10<=9       sonuç = False

## Mantıksal Operatörler

* and -----> ve operatörü: her iki durum doğru ise
    * (3 > 2) and (5 > 3)    ***True***
* or ------> veya operatörü: herhangi bir durum doğruysa True, hepsi yanlış ise False
    * (2 == 5) or (5 ==5)     ***True***
* not -------> değil operatörü: koşulların tersi alınması için kullanılır
    * a = 5, b = 2
    * not(a < 5  and b < 10)         ***False***

```python
# Determine if roots of polynomial ax^2 + bx + c = 0
# are real, repeated or complex using the
# quadratic formula x = (-b \pm \sqrt{b^2 - 4ac})/2a
a = 10
b = -234
c = 1984
discriminant = b**2 - 4*a*c
if discriminant > 0:
    print("Discriminant =", discriminant)
    print("Roots are real and distinct.")
elif discriminant < 0:
    print("Discriminant =", discriminant)
    print("Roots are complex.")
else:
    print("Discriminant =", discriminant)
    print("Roots are real and repeated.")
```

```python
import math
a = float(input("a katsayısı:"))
b = float(input("b katsayısı:"))
c = float(input("c katsayısı:"))
delta = pow(b, 2) - 4 * a * c
print("delta =", delta)
if delta < 0:
    print("sanal kök")
elif delta == 0:
    print("çakışık kök")
    x1 = -b / 2 * a
    print("kökler", x1)
else:
    x1 = (-b + math.sqrt(delta) - 2 * a)
    x2 = (-b - math.sqrt(delta) - 2 * a)
    print("ilk kök", x1, "ikinci kök", x2)
```

## While(-iken, olduğu sürece) Döngüsü

```python
"""Burada x değişkeni 5'ten küçük olduğu sürece ekrana 'Python işliyoruz.' yazısını yazdır."""
x = 1

while x < 5:
    print("Python işliyoruz.")
    x += 1  # x = x + 1
    # her seferinde x'i bir arttırdık yoksa sonsuz döngü olur.
```

```python
x = 1
while x <= 10:
    # 2'ye bölümünden kalan 1 ise 
    if x % 2 == 1:
        print(f'sayı tek: {x}')
    else:
        print(f'sayı çift: {x}')
    x += 1

print('bitti...')
```

```python
"""Dışarıdan negatif sayı girilinceye kadar sayıları toplar."""
sayi = 0
toplam = 0
n = 0
print("Bir sayı giriniz, negatif sayı döngüyü sonlandırır:")
while sayi >= 0:
    sayi = int(input())
    toplam += sayi
    n += 1
    
ortalama = toplam / n
print(" Sayıların Toplamı=", toplam)
print("Sayıların Ortalaması:", ortalama)
```

```python
# kullanıcıdan satır ve sutunu alnınan matrisin sadece köşegenlerini 1 geri kalanını 0 yazdır.
satir = int(input("satir :"))
sutun = int(input("sütun :"))
matris = [[0 for x in range(satir)]for y in range(sutun)]
for i in range(satir):
    for j in range(sutun):
        if satir > 2:
            matris[1][j] = 0
            matris[2][j]=0
            matris[i][0] = 1
            matris[i][-1] = 1
print(matris)
```

```python
# Belirlediğimiz soru adeti içinde girilen sayılardan maximum ve minimum bulma

soru = int(input("Soru adetini giririniz :"))
maxs = int(input("Sayıyı giriniz :"))

mins = maxs

for i in range(soru-1):
  a = int(input("Sayıyı giriniz :"))
  if a > maxs:
    maxs = a
  if a < mins:
    mins = a
  
print("Minimum Sayı :", mins, "Maksimum sayı :", maxs)
```

## Break, Continue, Pass

```python
# break methodu ile belirlediğimiz yere gelince devam etmez.
for a in "Python kursu":
    if a == "k":
        break
    print(a, end = "\t")
```

```python
# continue methodu ile belirlediğimiz yere gelince o kısmı atlar ve devam eder.
for a in "Python kursu":
    if a == "k":
        continue
    print(a, end = "\t")
```

```python
for a in "Python kursu":
    if a == "k":
        pass
    print(a, end = "\t")
```

```python
giris = 0
toplam = 0
print("Lütfen bir sayı giriniz, negatif sayılar döngüyü sonlandırır:")
while True:
    giris = int(input())
    if giris < 0:
        break # Döngüden çıkılıyor
    toplam += giris
print("Toplam =", toplam)
```

## Hata Yönetimi

- Şimdiye kadar yazdığımız tüm programlar  tek bir ortak varsayım üzerine kuruludur. Fakat bazen aksilikler olabiliyor. Örneğin, yazdığımız programı, kullanıcının bir sayı girmesi temeli üzerine ayarlamışken, kullanıcının her zaman sayı değerli bir veri gireceğinden emin olamayız. Bunun için programın hata vermemesi adına hata yakalama işlemlerini uyguluyoruz.
- Şimdi onları görelim.

```python
# Mesala şöyle bir program hata oluşturan bir kod yazalım ve görelim
```

```python
a = input("first number:")
b = input("second number:")

print(f"{a} / {b} is {int(a)/int(b)}")
```

```python
# Bu hataları yakalamak için try except bloklarını kullanırız. Yukarıdaki kodda sayıyı 0'a böldüğümüz zaman program hata verdi.
# Şimdi try except blok kullanarak tekrar deneyelim
```

```python
a = input("first number:")
b = input("second number:")

try:  
    # try kısmına hangi işlemin gerçekleşmesini istiyorsak onu yazıyoruz.
    print(f"{a} / {b} is {int(a)/int(b)}")

except ZeroDivisionError:
    # except ksımına beklediğimiz hata sonucunu yazıyoruz.
    print("Sıfıra bölünemez..!")
```

```python
"""
Try ... except çalışma prensibi :
try:
   hata verebileceğini bildiğimiz kodlar
except HataAdı:
   hata durumunda yapılacak işlem
"""
```

```python
# Kodumuzda birden çok hatalar gerçekleşebilir. Bunun için except bloklarını istediğimiz kadar kullanabiliriz.
```

```python
a = input("first number:")
b = input("second number:")

try:  
    print(f"{a} / {b} is {int(a)/int(b)}")

except ZeroDivisionError:
    print("Sıfıra bölünemez..!")
except ValueError:
    print("Lütfen sadece sayı giriniz")
```

```python
# Eğer istersek tek bir blok kısmında iki hatayı birden kodlayabiliriz.
```

```python
a = input("first number:")
b = input("second number:")

try:  
    print(f"{a} / {b} is {int(a)/int(b)}")

except (ZeroDivisionError, ValueError):
    print("ERROR..!")
```

```python
# Python'da hatalar hep genel bir hata yapısının altındadır. Örneğin işletim sistemindeki hatalar OS Exception altındadır.
# Bizde try except ile hata yakalarken tüm hataları tek tek yakalamak yerine Genel hata sınıfını yazabiliriz
```

```python
a = input("first number:")
b = input("second number:")

try:  
    print(f"{a} / {b} is {int(a)/int(b)}")
except Exception:
    # beklediğimiz hatalar Exception hata sınıfına ait hatalar // Programcı Hataları (Error) Program Kusurları (Bug) İstisnalar (Exception)
    print("ERROR..!")
```

## Try / except as 

- Program çalışma esnasında hata üretirken çıktıda hata türünün adıyla birlikte kısa bir hata açıklaması veriyor. Bu çıktı sayesinde daha kolay hatayı aratabiliyoruz.

```python
a = input("first number:")
b = input("second number:")

try:  
    print(f"{a} / {b} is {int(a)/int(b)}")
except ZeroDivisionError as hata:
    print(hata)    # Programın hata msjı
    print("sıfıra bölünemez")  # bizim yazdığımız hata msjı
```

## Try Except Else

- Aslındada biz bu else deyimini daha önce de ‘koşullu ifadeler’ konusunu işlerken görmüştük. Buradaki kullanımı da zaten hemen hemen aynıdır. Bize hataları teker teker inceleme imkanı sunar.

```python
try:
    a = int(input("bölünecek sayı: "))
    b = int(input("bölen sayı: "))
except ValueError:     # ilk girilen değerin sayı olma durumunu inceliyoruz
    print("Lütfen sadece sayı girin!")
else:        # eğer sayı ise bu sefer sıfıra bölünme durumuna bakıyoruz
    try:
        print(a/b)
    except ZeroDivisionError:
        print("0'a bölemezsiniz!")
```

## Try Except Finally

```python
# Bu yapının çalışma mantığı şu şekildedir;
```

```python
"""
try:
    ...bir takım işler...
except birHata:
    ...hata alınınca yapılacak işlemler...
finally:
    ...hata olsa da olmasa da yapılması gerekenler...
"""
```

```python
a = input("first number:")
b = input("second number:")

try:  
    print(f"{a} / {b} is {int(a)/int(b)}")

except (ZeroDivisionError, ValueError):
    print("ERROR..!")
finally:
    print("iyi çalışmalar")
```

# Fonksiyonlar

## Fonksiyon Nedir?

- Yazdığımız bir programda belli bir kod parçasını bir kaç yerde kullanma ihtiyacı duyduğumuzda fonksiyon oluşturmak işlerimizi kolaylaştırır. Böyle durumlarda sürekli kullanacak olduğumuz kod satırlarını fonksiyonlar içine alıp istediğimiz zaman çağırıp çalıştırabiliriz.

- Örneğin basit bir matematiksel işlem yapacağız ve bir çok yerde bu matematiksel işlemi tekrarlayacağız bu durumda bir fonksiyon oluşturup defalarca kullanabiliriz.

## Fonksiyonların Özellikleri

 - Her fonksiyonun bir ismi vardır ve fonksiyon_ismi() şeklinde kullanılırlar
 - Parantezlerin içerisine fonksiyonlara işlevsellik kazandıran parametreler yazılır.
 - Fonksiyonların bazı parametrelerinin varsayılan(default) değerleri vardır. Biz Değiştirmediğimiz sürece o değerler kullanılır => print(end=\n) örnek print fonksiyonu çalıştıktan sonra alt satıra geçmesini sağlayan end parametresine sahiptir.

## Fonksiyonların Görevleri

* Karmaşık işlemleri bir araya toplayarak, bu işlemleri tek adımda yapmamızı sağlamaktır. Fonksiyonlar çoğu zaman, yapmak istediğimiz işlemler için bir şablon vazifesi görür. Fonksiyonları kullanarak, bir veya birkaç adımdan oluşan işlemleri tek bir isim altında toplayabiliriz. Python’daki ‘fonksiyon’ kavramı başka programlama dillerinde ‘rutin’ veya ‘prosedür’ olarak adlandırılır. Gerçekten de fonksiyonlar rutin olarak tekrar edilen görevleri veya prosedürleri tek bir ad/çatı altında toplayan araçlardır.

```python
# print() fonksiyonunun görevi, kullanıcının girdiği parametreleri ekrana çıktı olarak vermektir.
# Burada print() fonksiyonu toplam dört adet parametre alıyor.
print("Tekirdağ", "Namık Kemal Üniversitesi", "Biyomedikal Mühendisliği", "2021")
```

- Çıktıda görüldüğü gibi print() fonksiyonu bazı işlemleri kendisi gerçekleştirdi.
- 1- İlk olarak kendisine verilen parametreleri(değerleri) ekrana bastırıyor 
- 2- Ekrana basarken sep=' ' parametresinin etkisiyle aralarına birer boşluk bırakıyor 
- 3- end='\n' parametresi sayesinde ekrana yazdırdıktan sonra otomatik olarak bir alt satıra geçiyor

```python
# Örneğin bir form kaydı oluşturalım ve her kaydı ekrana teker teker yazsın.
# Şimdiye kadar öğrendiklerimiz ile şu şekilde yapardık.
```

```python
isim     = "Rabia"
soy_isim = "Koç"
yas      = "22"
okul     = "Üniversite"

print("isim       :", isim)
print("soy_isim   :", soy_isim)
print("yas        :", yas)
print("okul       :", okul)
```

- Başka bir kişi için de kayıt oluşturmak istersek ne yapacağız?
- Tekrardan aynı işlemi yapmak zorundayız, bunun için böyle uğraşmak yerine bir fonksiyon oluştursak daha kolay olmaz mı? Örneğin; kayıd_olustur(Rabia, Koç, 22, Üniversite) diyip aldığı parametrelere göre daha hızlı kayıt işlemi gerçekleştirip ekrana yazmamız daha basit olacaktır. Tekrarlayan işlemleri yapmak için bizlerde fonksiyonları kullanıcağız.

## Fonksiyon Tanımlama ve Çağırma

* Fonksiyon tanımlamak için def komutunu kullanıyoruz.
* Fonksiyonları tanımlıyoruz ancak çağrılmadıkça fonksiyon bloğu içerisindeki kodlar çalıştırılmaz. Fonksiyonu çağırmak için fonksiyonun ismi ile birlikte parantezleri kullanmalıyız.

```python
def kayıd_olustur(isim, soyisim, yas, okul):     # def fonksiyon tanımlarken kullandığımız terimdir.
    print("-"*20)                 # Fonksiyonun içindeki işlemleri tanımlarken girintiye dikkat etmelisiniz.
    
    print("isim           : ", isim)
    print("soyisim        : ", soyisim)
    print("yaş            : ", yas)
    print("okul           : ", okul) 
    
    print("-"*20)
```

```python
# Fonksiyonumuzu tanımladık ve şimdi fonksiyonu çağırıp istediğimiz kaydı gerçekleştiricez.
kayıd_olustur("Rabia", "Koç", "22", "Üniversite")
```

```python
# başka bir kayıtta oluşturalım bakalım tekrarlıyacak mı kendisini
kayıd_olustur("Esra", "Turan", "21", "Üniversite")
```

### Rastgele sayıda isimsiz parametre yollama (Demet)

- Şimdiye kadar öğrendiğimiz çoğu fonksiyonun alabileceği max parametre sayısı belliydi. Daha fazlasını vermeye çalışırsak Python bize hata mesajı üretecekti. Ancak İstersek sınırsız sayıda parametre alabilen fonksiyon tanımlamamız mümkün. Gelin örnek üzerinde inceleyelim

```python
def sınırsız(*parametreler):
    print(parametreler)
```

```python
sınırsız(1, 2, 3, 4, 5, 6, 7, 8)
```

- (*) işareti bizim fonksiyonmuzun sınırsız sayıda parametre almasını sağlıyor NOT = pratikte sınırsız desekte teknik olarak alınabilecek parametre sayısı 256'dır

- Dikkat edilmesi gereken bir nokta iste fonksiyonun bize cevap olarak döndürdüğü sonuç demettir. Bu yüzden bir işlem yapmak istediğimizde demetlerde öğrendiğimiz konular geçerlidir

```python
def çarp(*sayılar):
    sonuç = 1
    for i in sayılar:
        sonuç *= i
    print(sonuç)
```

```python
çarp(2,5,7,4,3,5,7)
```

- (*) işaretini daha öncede öğrenmiştik. Orda yaptığı işlem karakterleri,listeleri birbirinden ayırmaktı. Fonksiyonlarda ise bu tam tersi. Örneklerde de görüldüğü üzere * tüm parametreleri birleştirip bir demet haline getiriyor

- (*) işaretinin yanındaki ismi istediğimiz gibi verebiliriz. Ancak genellikle *args kullanılır

### Rastgele Sayıda isimli parametre belirleme (Sözlük)

- bir önceki başlıkta fonksiyon tanımlarken rastgele sayıda isimsiz parametreleri belirledik. Şimdi de parametreleri yine sınırsız sayıda ancak isimli olarak belirleyebiliriz. Örnek üzerinde inceleyelim

```python
def fonksiyon(**bilgiler):
    print(bilgiler)
```

- Gördüğümüz gibi fonksiyonu çağırırken isimlendirdiğimiz parametreleri ve değerlerini sözlük yapısına getirdi. İşte bizde işlem yaparken sözlük tipinde geçerli olan kurallara uyarak işlem yapabiliriz

```python
def kayıt(**bilgiler):
    print("-"*20)

    for key, value in bilgiler.items():
        print("{}: {}".format(key, value))

    print("-"*20)
```

* işaretinin yanındaki ismi istediğimiz gibi verebiliriz. Ancak genellikle kwargs kullanılır.

- ( **ve ** * ) işaretlerini bir çok şekilde kullanabiliriz.
- Örnek olarak bir sözlük oluşturalım. Bunda türkçe kelimelerin karşılıkları olsun. Bizde istediğimiz kelimlerin sözlükte karşılığı olup olmadığını kontrol edelim

```python
def ceviri(*args, **kwargs):
    for kelime in args:
        if kelime in kwargs:
            print(f"{kelime} : {kwargs[kelime]}")
        else:
            print(f"{kelime} kelimesi sözlükte yok")
sözlük={"kitap"      : "book",
        "bilgisayar" : "computer",
        "sayı": "number"}
    
ceviri('araba', 'okul', 'bilgisayar', **sözlük)
```

- Dikkat edilmesi gereken bir noktada, ikisini birbiriyle kullanırken sırasının önemidir.
 - args(* *) her zaman kwargs(** *) önce gelmelidir. Yoksa Python hata üretecektir

## Fonksiyonların Kapsamı ve global deyimi

```python
x = 0

def fonk():
    x = 1
    return x

# elimizde böyle bir kod var. fonksiyon: içindeki x fonksiyon dışındakini etkiler mi 
```

```python
print('fonksiyon içindeki x: ', fonk())
print('fonksiyon dışındaki x: ', x)
```

- Görüldüğü üzere fonksiyon içindeki x dışarıyı etkilemedi. Bunun sebebi Python'daki namespace(isim alanı) kavramıdır. Peki isim alanı nedir. Değişkenlerin, fonksiyonların ve sonraki konularda göreceğimiz sınıfların birer kapsama alanı vardır. Buna namespace denir. Dolayısıyla Pythonda her nesnenin geçerli ve etkin olduğu bir alan vardır. Mesela yukarıda ki fonksiyonumuzun dışındaki x ana namespace alanında yer alan global bir değişkendir. Fonksiyonun içinde ki x ise fonk() isim alanı içinde yer alan local bir değişkendir. isimleri aynı olsada isim alanları farklı olduğunda farklı birer değişkenlerdir.

```python
x = []
print('x\'in ilk hali:', x)

def değiştir():
    print('x\'i değiştiriyoruz...')
    x.append(1)
    return x

değiştir()
print('x\'in son hali: ', x)
```

- fonksiyon içinde ki x'e yapılan değişiklik genel olarak bir değişime sebep oldu. Yukarda bunun tam tersini söylemiştik bunun sebebi nedir acaba ? Pythonda herhangi bir nesneye gönderme yaptığımızda öncelikle nesneyi mevcut isim alanında arar. Eğer aranan nesne mevcut isim alanında yok ise yukarı doğru tüm isim alanlarını kontrol eder

- Karakter dizileri gibi değiştirilemeyen dizileri, local isim alanlarında değiştiremiceğiniz gibi yeniden tanımlayamazsınızda. Aynı kural değiştirilebilir nesneler içinde geçerlidir. -- Tanımlamak ve yeniden tanımlamak farklı şeylerdir.

```python
isim_listesi = []

def fonk():
    isim_listesi += ['Rabia Koç', 'Yusuf Akyol']   # += yerinde sadece = koyarsak hata vermez
    return isim_listesi

print(fonk())
```

- Yukarıda bahsettiğimiz kural burda geçerli. İsim listesi daha önceden global isim alanında tanımlanmıştı. Ancak local isim alanında aynı değişken tekrar tanımlanmaya çalıştığı için hata verdi. isim_listesi = ['Rabia Koç', 'Yusuf Akyol'] şeklinde bir tanımlama yapmış olsaydık hata almayacaktık. Çünkü local isim alanında hiç var olmayan bir değişken tanımlamış oluyoruz

* Bu bahsettiğimiz hatayı çözmenin bir yolu mevcut. o da global deyimi. Örnek üstünde görelim

```python
isim_listesi = []

def fonk():
    global isim_listesi
    isim_listesi += ['Rabia Koç', 'Yusuf Akyol']  
    return isim_listesi

print(fonk())
```

- Burada kullandığımız global değimi isim_listesi değişkeninin global alana taşınmasını sağlıyor. Bu sayede üzerinde değişiklik yapabiliyoruz. Ancak bu deyimi kullanırken dikkatli olmalıyız. Eğer dikkatsiz kullanırsak programımızın hatalı çalışmasına yol açabilir.

## enumerate() Fonksiyonu

* Eğer yazdığımız bir programda numaralandırmaya ilişkin işlemler yapmanız gerekiyorsa bu fonksiyonu kullanırız.
* Enumerate methodu iki parametre alır. Birinci parametre itere edilecek objedir yani bir iterabledır. Örneğin bir list, tuple, yada string gibi. İkinci parametre ise start parametresidir ve indexlenmenin kaçtan başlayacağını belirler. Bu parametre opsiyoneldir, istersek kullanmayabiliriz ama bu durumda index 0'dan başlayacaktır.

```python
isimler = ['Oğuzhan', 'Enes', 'Selim']
counter = 1
for isim in isimler:
  print(counter, isim)
  counter += 1
```

* Bu kod bloğunda, for döngüsünün dışında bir counter(sayaç) tanımladık. Daha sonra bu counterın değerini her döngü döndüğünde bir arttırdık. Bu sayede aslında listenin sıra numaralarını yazdık.

```python
# Peki bunu nasıl daha güzel bir şekilde yapabiliriz?
```

```python
isimler = ['Oğuzhan', 'Enes', 'Selim']
for index, isim in enumerate(isimler, start=1):
  print(index, isim)
```

- Eğer ki print(enumerate(isimler, start=1)) kodunu çalıştırırsanız ve enumeratein ne yaptığını görmek için yazdırırsanız, hüsrana uğrayabilirsiniz çünkü geriye bir enumerate objesi dönecektir. Beklediğimiz şekilde bir liste ve bunun index numaraları dönmeyecektir.

### Enumerate nasıl indexliyor?

```python
print(list(enumerate(isimler)))
```

* Enumerate, listenin içerisinde ki isimleri aldı daha sonra yanlarına bir index numarası koydu ve bunları bir tuple haline getirip dedi ki bu tuple’ın ilk elemanı index numarası ikinci elemanı ise elemanın ta kendisi olsun.

```python
# [(0, 'Oğuzhan'), (1, 'Enes'), (2, 'Selim')]
# Bunun ise bize for döngüsünde harika bir avantajı oldu!
```

```python
# for index, isim in enumerate(isimler, start=1):
```

* For döngüsünün sağ tarafında ki kısım içi tuple dolu bir liste döndürdüğüne göre her döngüde sol taraf bir tuple olacaktır! Biz hem index numarasını hemde bu değeri almak istediğimiz için forun hemen sağına bir adet index bir adet de isim adında değişken tanımladık. Bu sayede her for döngüsü döndüğünde index değişkeni üzerinde o dönen tupleın index numarasını, isim değişkeni üzerinden de tupleın içinde ki değeri alabiliyoruz.

### Tuple ve String ile Enumerate Kullanımı

-Sonuç olarak Tuple da String de liste gibi bir iterable olduğu için kullanım şekli fark etmeksizin enumeratein birinci parametresine getirilebilirler.

```python
isim = 'Rabia' 
for index, isim in enumerate(isim, start=1):
    print(index, isim)
```

* 1- isim adında bir değişken tanımladık.

```python
maaslar = [1000, 2000, 3000, 4000, 5000]

def maas_ust(x):
    print(x*10/100 + x)

def maas_alt(x):
    print(x*20/100 + x)

for i in maaslar:
    if i >= 3000:
        maas_ust(i)
    else:
        maas_alt(i)
```

```python
# ebob bulma:
def obeb(sayi1, sayi2):
    for i in range(int((sayi1 + sayi2) / 2), 0, -1):
        if sayi1 % i == 0 and sayi2 % i == 0:
            break
    return i


print(obeb(36, 78))
```

```python
# bir dizideki elemanların karesini al sonra bunları ortalamasını al ve kökünü al.
import math
dizi = [3, 5, -7, 8, 4]
yd = []
for i in (dizi):
    yd.append(int(math.pow(i, 2)))
    print(yd)
ort = (yd[0]+yd[1]+yd[2]+yd[3]+yd[4])/len(yd)
print(math.sqrt(ort))
```

Sınıflar nesne üretmemizi sağlayan veri tipleridir. 1 kere daire nesnesini yarattığımız zaman diğer tüm dairelerimiz için bu nesneyi kullanıp istediğimiz işlemleri kolay bir şekilde yapabiliriz.

```python
# self1: anahtar sözcüğü kullanarak sınıfın methodlarına python ile erişebilir.
# toString içine self yazarsak toStringi Circle classına bağlar.
class Circle:
    # böyle yaparak pi değerini tüm Circle içerisinde erişebilirz.
    global pi 
    pi=3.14
    
    # constructor: yani bir dairenin temeldeki özelliklerini, temel niteliklerini tanımlamak
    # clasların içinde kullandığımız fonk. method olarak adlandırırız.
    def __init__(self, name, r):
        self.class_name = name   
        # self sözcüğü ile Circle class'ına bağlamış oldu.
        self.class_r = r
        
# dairenin içindeki class'lara erişmek için
# eğer self keyword'ü buraya koymazsak toString methodunu Circle class'ı ile bağlayamayız.
    def toString(self):            
        print(self.class_name)
        print(self.class_r)
        
# dairemizin alanını bulmak için
    def findArea(self):  
        area = pi * self.class_r * self.class_r  
        print(area)
        
# class_name ismini değiştirmek için buraya "name" input olarak girdik.
    def setName(self, name): 
        self.class_name = name 

# class_name değiştikten sonra ne olduğuna bakmak için
    def getName(self):  
        print(self.class_name)
        return self.class_name

# bir tane method yazdık ve içeriğine daha karar vermedik pass yazarsak hiçbir hata çıkmıcak.
    def emp(self): 
        pass
    
# bir obje tanımladık.
c1 = Circle("s1", 2)

c1.toString()
```

Ancak biz çoğu zaman dosya imlecini (file) dosyanın herhangi bir yerine götürmek isteyebiliriz. Bunun için Pythondaki seek() fonksiyonunu kullanacağız. * ***Şu anda hiçbir işlem yapmadığımız için tell() fonksiyonu dosyanın en başında (0. byteda) olduğumuzu söyledi. Peki bir dosya imlecini dosyanın 20.byte’ına götürmek için ne yapacağız ? Bunun için de seek() fonksiyonunu kullanacağız.***

```python
def not_hesapla(satır):
    
    satır = satır[:-1]  # /n dışında bütün satırları alıyor. sondaki /n'i silmiş oluyor.

    liste = satır.split(",") # her bir elemanı ayrı ayrı böler

    isim = liste[0] 

    not1 = int(liste[1])

    not2 = int(liste[2])

    not3 = int(liste[3])

    son_not = not1 * (3/10) + not2 * (3/10) + not3 * (4/10)

    if (son_not >= 90):

        harf = "AA"
    elif (son_not >= 85):
        harf = "BA"
    elif (son_not >= 80):
        harf = "BB"
    elif (son_not >= 75):
        harf = "CB"
    elif (son_not >= 70):
        harf = "CC"
    elif (son_not >= 65):
        harf = "DC"
    elif (son_not >= 60):
        harf = "DD"
    elif (son_not >= 55):
        harf = "FD"
    else:
        harf = "FF"

    return isim + "------------------> " + harf + "\n"  # her bir satırın değerini bulup dönmüş olur.

# önce dosya okuma işlemi yaptık
with open("dosya.txt", "r", encoding = "utf-8") as file:

    eklenecekler_listesi = []

    for i in file:
        # her bir satıra karşılık gelen listeye eklenir.
        eklenecekler_listesi.append(not_hesapla(i))
        
# notlar dosyasında oluşmuş oldu.
    with open("notlar.txt", "w",encoding = "utf-8") as file2:

        for i in eklenecekler_listesi:
            file2.write(i)
```

## İleri Toplulaştırma İşlemleri (Aggregate, Filter, Transform, Apply)

```python
import pandas as pd
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10, 23, 33, 22, 11, 99],
                   'degisken2': [100, 253, 333, 262, 111, 969]},
                   columns = ['gruplar', 'degisken1', 'degisken2'])
df
```

* ***AGGREGATE***

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").aggregate(["min", np.median, "max"])
# veri setini gruplara göre böl sonra liste formatında beklediğimiz fonksiyonları girdik
# pandasın içinde yer aldığını bildiğimiz fonksiyonları ya tırnak içine alırız ya da almayız.
# diğer fonksiyonları da yer aldığı kütüphaneye bağlarız.(np.median)
```

```python
df.groupby("gruplar").aggregate({"degisken1": "min", "degisken2": "max"})
# 2 değişken içinde 2 farklı bir istatiksel uygulamak için
```

* ***FİLTER***

```python
import pandas as pd
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10, 23, 33, 22, 11, 99],
                   'degisken2': [100, 253, 333, 262, 111, 969]},
                   columns = ['gruplar', 'degisken1', 'degisken2'])
df
```

```python
def filter_func(x):
    return x["degisken1"].std() > 9
```

```python
df.groupby("gruplar").std()
```

```python
df.groupby("gruplar").filter(filter_func)
```

* ***TRANSFORM***

```python
import pandas as pd
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10, 23, 33, 22, 11, 99],
                   'degisken2': [100, 253, 333, 262, 111, 969]},
                   columns = ['gruplar', 'degisken1', 'degisken2'])
df
```

## Pivot Tablolar

```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
titanic.groupby(["sex", "class"])[["survived"]].aggregate("mean").unstack() 
# daha düzenli tablo yapa "unstack()" 
```

* ***pivot ile table***

```python
titanic.pivot_table("survived", index = "sex", columns = "class") 
# index olatak girdiğimiz yer satır olarak değerlendiriyor.
# bu şekilde de üstteki olan şey bulunur.
```

```python
titanic.age.head()
```

```python
# bu sürekli değişkeni bir kategorik değişkene çevirip örneğin üst 
# sınıflı bir kategorik değişkene çevirip bu kategorik değişkenin 
# sınıflarını da pivot tabloya boyut olarak ekliyeyim.
```

```python
age = pd.cut(titanic["age"], [0, 18, 90])
age.head(10)
#titanic verisinin içerisindeki age değişkenini böl.
#şu değerlere göre böl.
```

```python
# ortaya çıkan bu sınıfları bir boyut olarak nasıl pivot table ye ekleyebiliriz
```

```python
titanic.pivot_table("survived", ["sex", age], "class")
#age yi dışardan aldık.
```
